```python
import statsapi
import json
import requests
from datetime import datetime
today = datetime.today().strftime('%Y-%m-%d')
print(today)
opening_day = "2024-03-28"
```

```python
players = [{"Will Smith", "Matt Olson", "Nolan Gorman", "Gunnar Henderson", "José Ramírez", "Fernando Tatis Jr.", "Teoscar Hernández", "Jorge Soler"},
           {"Yainer Diaz", "Bryce Harper", "Marcus Semien", "Rafael Devers", "Corbin Carroll", "Kyle Tucker", "Aaron Judge"},
           {"Cal Raleigh", "Vladimir Guerrero Jr.", "Ozzie Albies", "Corey Seager", "Austin Riley", "Julio Rodríguez", "Cody Bellinger", "Anthony Santander", "Marcell Ozuna"},
           {"Adley Rutschman", "Pete Alonso", "Ketel Marte", "Bobby Witt Jr.", "Manny Machado", "Nick Castellanos", "Christopher Morel", "Juan Soto", "Shohei Ohtani"},
           {"Spencer Torkelson", "Brandon Drury", "Francisco Lindor", "Max Muncy", "Adolis García", "Kyle Schwarber"}
          ]

teams = ["Jake/Sam", "Jamie/Jason", "Vitus/Luke", "Griffin/Matt", "Owen/Daniel"]
homers = {}
for player_set,team in zip(players, teams):
    homers[team] = {}
    for player in player_set:
        try:
            players_data = statsapi.get('sports_players', {'season': 2024, 'gameType': 'W'})
                
            # Fetch player ID
            player_id = next(x['id'] for x in players_data['people'] if x['fullName'] == player)
            stats_json = statsapi.player_stat_data(player_id, "hitting", type="season", sportId=1)
        
        
            # stats_json = parse_stats(stats)
            # print(stats_json)
            # print(f"{player}: {stats_json}")
            
            # Access home runs if available
            if 'stats' in stats_json and len(stats_json['stats']) > 0:
                # Loop through stats to find the 'hitting' group if there are multiple stat groups
                for stat_group in stats_json['stats']:
                    if stat_group['type'] == 'season' and stat_group['group'] == 'hitting':
                        home_runs = stat_group['stats'].get('homeRuns', 'N/A')
                        homers[team][player] = home_runs
                        # print(f"{player}:{home_runs}")
                        
            else:
                print(f"{player}: No statistics available")
        except StopIteration:
            print(f"Player ID not found for {player}")
        except Exception as e:
            print(f"An error occurred for {player}: {e}")

print(homers)
```

```python
for team in homers:
    total = 0
    for player in homers[team]:
        if player != "total":  # Exclude any existing "total" entries
            total += homers[team][player]
    homers[team]["total"] = total
homers
```

```python
import csv

file = "player_home_runs.csv"

with open(file, mode='w', newline='') as file:
    writer = csv.writer(file)
    # Write header
    writer.writerow(["Team", "Player", "Home Runs"])
    
    # Write data rows
    for team, players in homers.items():
        for player, home_runs in players.items():
            writer.writerow([team, player, home_runs])

print(f"Data written to {file}")
```
